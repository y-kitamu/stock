```python
def calc_costs(data_str):
    data = np.reshape(data_str.splitlines(), (-1, 7))
    buys = data[data[:, 1] == '買付']
    sells = data[data[:, 1] == '売付']

    total_cost = 0
    num_buy = 0

    for trade in buys:
        total_cost += float(trade[5]) * int(trade[6])
        num_buy += int(trade[6])

    total_sell = 0
    num_selled = 0

    for trade in sells:
        total_sell += float(trade[5]) * int(trade[6])
        num_selled += int(trade[6])

    print("cost = {}, sell = {}".format(total_cost / num_buy, total_sell / num_selled))
```
